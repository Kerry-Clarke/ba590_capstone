### Combine separate dataframes into a Master File 

```python
# append df17 to the end of the df18
df = df18.append(df17, sort=False)
df.head()
```

```python
# verify that appropriate columns were merged
# also check the number of rows (entries)
df.info()
```

```python
# append df16
df = df.append(df16, sort=False)
df.head()
```

```python
# verify columns and entries 
df.info()
```

```python
# append df15
df = df.append(df15, sort=False)
df.info()
```

```python
# append df14
df = df.append(df14, sort=False)
# verify columns and entries 
df.info()
```

```python
# drop empty column from master dataframe 
df = df.drop(['Unnamed: 39'], axis=1)
```

```python
# export master dataframe to csv file 
df.to_csv('data/admission_master.csv', index=0)
```
