```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
plt.style.use('ggplot')
```

```python
# specific population is the students who meet the following criteria:
    # - are/were undergraduates
    # - the "term indicator" = Fall
    # - classified as FR (Freshmen)
    # - their cohort was 201409F, 201509F, 201609F, 201709F or 201809F
    # - were/are degree-seeking
# This population excludes any students whose initial term was not in the Fall (transfers) 

df = pd.read_excel('Census.xlsx', sheet_name='2014-2018')
df.head()
```

```python
df.columns, df.info()
```

```python
# gender 
fig = plt.figure(figsize=(8,6))
df['SEX'].value_counts().plot(kind='bar')
plt.title('Student Gender 2014-2018')
plt.xticks(rotation='horizontal')
plt.show()
plt.savefig('images/gender.pdf')
```

```python
# ethnicity
fig = plt.figure(figsize=(8,6))
df['Ethnicity'].value_counts().plot(kind='bar')
plt.title('Student Ethnicity 2014-2018')
plt.show()
plt.savefig('images/ethnicity.png')
```

```python
# residency
fig = plt.figure(figsize=(8,6))
df['Residency'].value_counts().plot(kind='bar')
plt.title('Student Residency Status 2014-2018')
plt.xticks(rotation='horizontal')
plt.show()
plt.savefig('images/residency.png')
```

```python
# residency
fig = plt.figure(figsize=(8,6))
df['Athlete - Indicator'].value_counts().plot(kind='bar')
plt.title('Student Athlete Status 2014-2018')
plt.xticks(rotation='horizontal')
plt.show()
plt.savefig('images/athlete.png')
```

```python
# state
fig = plt.figure(figsize=(20,5))
df['STATE'].value_counts().plot(kind='bar')
plt.title('Student Home States 2014-2018')
plt.xticks(rotation='horizontal')
plt.show()
plt.savefig('images/state.png')
```

```python
# admission rating 
fig = plt.figure(figsize=(8,6))
df['Admission Rating'].value_counts().plot(kind='bar')
plt.title('Student Admission Rating 2014-2018')
plt.show()
plt.savefig('images/admission rating bar.png')
```

```python
# admission rating
fig = plt.figure(figsize=(8,6))
df['Admission Rating'].plot(kind='hist')
plt.title('Student Admission Rating 2014-2018')
plt.show()
plt.savefig('images/admission rating histogram.png')
```

```python
fig = plt.figure(figsize=(8,6))
df['First_Gen'].value_counts().plot(kind='bar')
plt.title('First Generation Students 2014-2018')
plt.xticks(rotation='horizontal')
plt.show()
plt.savefig('images/first_gen.png')
```

```python
# colleges
fig = plt.figure(figsize=(8,6))
df['College Final'].value_counts().plot(kind='bar')
plt.title('Colleges 2014-2018')
plt.show()
plt.savefig('images/college.png')
```

```python
# majors
fig = plt.figure(figsize=(20,5))
df['Major Complete'].value_counts().plot(kind='bar')
plt.title('Student Majors 2014-2018')
plt.show()
plt.savefig('images/majors.png')
```

```python
# HS GPA and Fall Semester Freshman Year GPA 
fig = plt.figure(figsize=(15,5))
plt.tight_layout()

ax1 = fig.add_subplot(1,2,1)
ax2 = fig.add_subplot(1,2,2)

ax1.hist(df['High School GPA'])
ax2.hist(df['GPA (end of 1st semester FR year)'])

ax1.set_title('High School GPA 2014-2018')
ax2.set_title('Fall Semester Freshman Year GPA 2014-2018')

ax1.set_ylabel('Number of Students')
ax2.set_ylabel('Number of Students')

plt.savefig('images/gpa.png')
```

```python
# ACT and SAT Scores
fig = plt.figure(figsize=(15,5))
plt.tight_layout()

ax1 = fig.add_subplot(1,2,1)
ax2 = fig.add_subplot(1,2,2)

ax1.hist(df['ACT'])
ax2.hist(df['SAT Comp (new)'])

ax1.set_title('Student ACT Scores 2014-2018')
ax2.set_title('Student Total SAT Scores 2014-2018')

ax1.set_ylabel('Number of Students')
ax2.set_ylabel('Number of Students')

plt.savefig('images/standard_tests.png')
```

```python
# SAT Score Breakdown
fig = plt.figure(figsize=(30,20))
plt.tight_layout()

ax1 = fig.add_subplot(2,3,1)
ax2 = fig.add_subplot(2,3,2)
ax3 = fig.add_subplot(2,3,3)

ax1.hist(df['SAT Comp (new)'])
ax2.hist(df['SAT Critical Reading (new)'])
ax3.hist(df['SAT Math (new)'])

ax1.set_title('Student Total SAT Scores 2014-2018')
ax2.set_title('Student SAT Critical Reading Scores 2014-2018')
ax3.set_title('Student SAT Math Scores 2014-2018')

ax1.set_ylabel('Number of Students')
ax2.set_ylabel('Number of Students')
ax3.set_ylabel('Number of Students')

plt.savefig('images/sat_breakdown.png')
```
